```python
"""
7. Build Order: You are given a list of projects and a list of dependencies (which is a list of pairs of
projects, where the second project is dependent on the first project). Ail of a project's dependencies
must be built before the project is. Find a build order that will allow the projects to be built. If there
is no valid build order, return an error.
Input:
p r o j e c t s : a, b, c, d, e, f
dependencies: (a, d), ( f , b ) , (b, d ) , ( f , a ) , (d, c)
Output: F, e, a, b, d, c
"""

projects = ["a", "b", "c", "d", "e", "f"]
dependencies = [("a","d"), ("f","b"), ("b","d"), ("f","a"), ("d","c")]


class Project:
    def __init__(self, name):
        self.dependencies = []
        self.name = name
    
    def isReady(self):
        return len(self.dependencies) == 0
    
    def removeDependency(self, target):
        if(target in self.dependencies):
            self.dependencies.remove(target)
        
    def addDependency(self, neededToBuild):
        self.dependencies.append(neededToBuild)
        
    def getDependencyList(self):
        return self.dependencies
    
    def name(self):
        return self.name

projectObjects = []
for project_ID in projects:
    obj = Project(project_ID)
    for edge in dependencies:
        if edge[1] == project_ID:
            obj.addDependency(edge[0])
    projectObjects.append(obj)
    

def getOrder(projects):
    orderList = []
    '''Breakout determines when to give up looking for build solutions'''
    breakout = len(projects) * 100
    counter = 0
    while(len(orderList) != len(projects) and counter < breakout):
        for proj in projects:
            if(proj.name not in orderList):
                if proj.isReady():
                    orderList.append(proj.name)

        for projectName in orderList:
            for proj in projects:
                proj.removeDependency(projectName)
                
        counter = counter + 1
    
    if(len(orderList) == len(projects)):
        return orderList
    return None

print(getOrder(projectObjects))
```
